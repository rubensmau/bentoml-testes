```python
import pandas as pd
import numpy as np
import json
```

```python
def processa_colunas(df):
    
    #####   traduz campos das colunas categ√≥ricas
    coluna = "employment__type"
    dic_troca = {'labourer': 'operario', 'professional':'nivel medio', 'unemployed':'desempregado', 'management': 'nivel alto'}
    df.replace({coluna: dic_troca},inplace=True)      
    
    coluna = "finance__accounts__checking__balance"
    dic_troca = {'no_account':'sem conta', 'low' : 'baixa', 'negative' : 'negativa', 'high' : 'alta'}
    df.replace({coluna: dic_troca},inplace=True)
    
    coluna= 'finance__accounts__savings__balance'
    dic_troca = {'very_high': 'muito_alta' ,'low': 'baixa' , 'high': 'alta' , 'medium': 'media',  'no_account': 'sem conta'}
    df.replace({coluna: dic_troca},inplace=True)
    
    coluna= 'contact__has_telephone'
    dic_troca = {False: 'Nao' ,True: 'Sim' }
    df.replace({coluna: dic_troca},inplace=True)
    
    coluna= 'finance__repayment_history'
    dic_troca = {'very_poor': 'muito pobre' ,'good': 'boa','poor': 'pobre' }
    df.replace({coluna: dic_troca},inplace=True)
    
    coluna=  'finance__other_assets'
    dic_troca = {'real_estate': 'imovel' ,'car_or_other': 'carro ou outro','life_insurance': 'seguro de vida', 'none' : 'Nao'  }
    df.replace({coluna: dic_troca},inplace=True)

    coluna=  'credit__purpose'
    dic_troca = {  'new_car': 'carro novo' ,'electronics': 'eletronico','re-training': 'requalificacao', 'appliances' : 'eletrodomestico',
            'education' : 'educacao',   'business'   : 'empresa',       'repairs': 'reformas',         'furniture': 'mobiliario',
             'used_car' :'carro_usado' ,     'other' : 'outro' }
    df.replace({coluna: dic_troca},inplace=True)
    
    coluna=  'residence__type'
    dic_troca = {'own': 'proprio' ,'rent': 'aluguel','free': 'gratis'  }
    df.replace({coluna: dic_troca},inplace=True)
    
    ##  multiplica a quantia por 6 --- converte para real
    df['credit__amount'] = df['credit__amount']*6
    
    ## dropa colunas
    df.drop(to_drop, axis=1,inplace=True)
    
    ### troca nome das colunas
    df.rename(columns=dic_traducao,inplace=True)
    return df
```

```python
### COLUNAS PARA DROP 
to_drop = ['employment__permit',]
```

```python
dic_traducao = {    'credit__coapplicant' : 'credito_coaplicante',                  'credit__guarantor'  : 'credito_fiador',
         'finance__credits__other_stores' : 'financiamento_outro_lugar',           'employment__permit'  : 'carteira_trabalho',
             'employment__type'           : 'tipo_emprego',   'finance__accounts__checking__balance'     : 'conta_corrente_valor',
     'finance__credits__this_bank'        : 'outros_creditos_aqui',    'finance__credits__other_banks   ': 'credito_outro_lugar',
                'personal__num_dependents': 'numero_dependentes',  'finance__accounts__savings__balance' : 'investimentos_valor',
                    'residence__duration' : 'duracao_residencia',                        'credit__amount': 'valor_solicitado',
                 'contact__has_telephone' : 'tem_telefone',                       'employment__duration' : 'duracao_emprego', 
                       'credit__duration' : 'duracao_credito',              'finance__repayment_history' : 'historico_credito' ,
                  'finance__other_assets' : 'financiamento_outros_bens',               'credit__purpose' : 'proposito_credito',
                        'residence__type' : 'tipo_residencia',                'credit__installment_rate' : 'receita_disponivel' 
              }
```

```python
dic = []
with open("train_data.json") as f:
    for line in f:
        dic.append(json.loads(line))
df = pd.DataFrame.from_dict(dic) 
df = processa_colunas(df)

#Lista todas as colunas
#df.columns
```

```python
###  LE TRAIN LABELS
dic = []
with open("train_label.json") as f:
    for line in f:
        dic.append(json.loads(line))
df_label = pd.DataFrame.from_dict(dic)      
#df_label.head(10)
coluna=  'credit__default'
dic_troca = {  False: 0 ,True: 1,}
# troca os valores das colunas
df_label.replace({coluna: dic_troca},inplace=True)
# troca o nome da coluna
df_label.rename(columns={'credit__default': 'deu_default'},inplace=True)
```

```python
###  Junta os dois dataframes em um apenas
df_train = pd.concat([df_label, df], axis=1)
df_train.head(1)

df_train.to_csv('train.csv')
```
